将你的 微信 头像（或者微博头像）右上角加上红色的数字，类似于微信未读信息数量那种提示效果。 类似于图中效果

```python
from PIL import Image, ImageDraw, ImageFont

def add_num(img):
    draw = ImageDraw.Draw(img)
    myfont = ImageFont.truetype('C:/windows/fonts/Arial.ttf', size=100)
    fillcolor = "#ff0000"
    width, height = img.size
    draw.text((width-180, 0), '99+', font=myfont, fill=fillcolor)
    img.save('result.jpg')
    img.show()
    return 0


if __name__ == '__main__':
    image = Image.open('0000.jpg')
    add_num(image)
```
